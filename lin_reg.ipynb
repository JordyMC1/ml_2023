```python
x=[1.0,1.6,3.4,4.0,5.2]
y=[1.2,2.0,2.4,3.5,3.5]
print(x)
print(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.array([1.0,1.6,3.4,4.0,5.2])
y = np.array([1.2,2.0,2.4,3.5,3.5])
plt.title("Plotting 1-D array")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color = "red", marker = "o", label = "Array elements")
plt.legend()
plt.show()
```

```python
sumax=sum(x)
print(sumax)
sumay=sum(y)
print(sumay)
sumaxy=sum(x*y)
print(sumaxy)
sumaxx=sum(x*x)
print(sumaxx)
n=5
a=((n*(sumaxy))-((sumax)*(sumay)))/((n*sumaxx)-(sumax*sumax))
print(a)
b=((sumay*sumaxx)-(sumaxy*sumax))/((n*sumaxx)-(sumax*sumax))
print(b)
y1=a+(b*x)
print(y1)
```
